```python
partners_tenure = tdf2[['Partner','tenure']].groupby(['Partner']).agg('median').reset_index()
partners_pay_mothly = tdf2[['Partner','MonthlyCharges']].groupby(['Partner']).agg('median').reset_index()
partners_pay_total = tdf2[['Partner','TotalCharges']].groupby(['Partner']).agg('median').reset_index()
df = [partners_pay_mothly,partners_pay_total,partners_tenure]
for i in df:
    i.loc[i['Partner'] == 'No', 'Partner'] = 'No Partner'
    i.loc[i['Partner'] == 'Yes', 'Partner'] = 'With Partner'
    i.rename(columns = {'Partner':'Category'}, inplace = True)
    
partners_pay_mothly
```

```python
Dependents_pay_monthly = tdf2[['Dependents','MonthlyCharges']].groupby(['Dependents']).agg('median').reset_index()
Dependents_pay_total = tdf2[['Dependents','TotalCharges']].groupby(['Dependents']).agg('median').reset_index()
Dependents_tenure = tdf2[['Dependents','tenure']].groupby(['Dependents']).agg('median').reset_index()
df = [Dependents_pay_total,Dependents_pay_monthly,Dependents_tenure]
for i in df:
    i.loc[i['Dependents'] == 'No', 'Dependents'] = 'No Dependents'
    i.loc[i['Dependents'] == 'Yes', 'Dependents'] = 'With Dependents'
    i.rename(columns = {'Dependents':'Category'}, inplace = True)
    
Dependents_tenure
```
